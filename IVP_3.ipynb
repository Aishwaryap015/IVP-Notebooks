ðŸ“Œ Overview

Histogram Equalization is a technique used to enhance the contrast of images.

It redistributes pixel intensity values across the full range (0â€“255).

Enhances visibility in low-contrast images.

Equivalent in MATLAB/Octave:

imhist() â†’ compute histogram

histeq() â†’ histogram equalization

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Upload image from local machine
uploaded = files.upload()

# Read the uploaded image (grayscale)
for filename in uploaded.keys():
    img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# --- Step 3: Compute Histogram & Apply Equalization ---

# Compute histogram of original image
hist_original = cv2.calcHist([img], [0], None, [256], [0, 256])

# Apply histogram equalization
equalized_img = cv2.equalizeHist(img)

# Compute histogram of equalized image
hist_equalized = cv2.calcHist([equalized_img], [0], None, [256], [0, 256])

# --- Quick Results ---
print("âœ… Step 3 Completed")
print("Original Image Shape:", img.shape)
print("Equalized Image Shape:", equalized_img.shape)

# Show quick preview of equalized image
plt.imshow(equalized_img, cmap='gray')
plt.title("Equalized Image (Preview at Step 3)")
plt.axis('off')
plt.show()
```

```python
# --- Step 4: Visualize Histograms & Images ---

plt.figure(figsize=(16,10))

# Original Image
plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

# Equalized Image
plt.subplot(2,2,2)
plt.imshow(equalized_img, cmap='gray')
plt.title("Equalized Image")
plt.axis('off')

# Histogram of Original Image
plt.subplot(2,2,3)
plt.plot(hist_original, color='black')
plt.title("Original Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

# Histogram of Equalized Image
plt.subplot(2,2,4)
plt.plot(hist_equalized, color='black')
plt.title("Equalized Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.suptitle("Histogram Equalization - Visualization", fontsize=16)
plt.tight_layout()
plt.show()

print("âœ… Step 4 Completed: Visualizations Generated")
```

```python
# --- Step 5: Before & After Comparison ---

plt.figure(figsize=(20,10))

# Before Equalization
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Before Equalization")
plt.axis('off')

# After Equalization
plt.subplot(1,2,2)
plt.imshow(equalized_img, cmap='gray')
plt.title("After Equalization")
plt.axis('off')

plt.suptitle("Contrast Enhancement using Histogram Equalization", fontsize=18)
plt.show()

print("âœ… Step 5 Completed: Before & After Comparison Displayed")
```

```python
# --- Step 6: Save & Download Equalized Image ---

# Save equalized image as a new file
output_filename = "equalized_image.png"
cv2.imwrite(output_filename, equalized_img)

# Download the saved file to your laptop
from google.colab import files
files.download(output_filename)

print(f"âœ… Step 6 Completed: File saved as {output_filename} and ready for download")
```

```python
# ðŸ“ˆ Overlayed Histogram Graph
plt.figure(figsize=(8,6))
plt.plot(hist_original, color='red', label="Original Histogram")
plt.plot(hist_equalized, color='blue', label="Equalized Histogram")
plt.title("Overlayed Histogram Comparison")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.legend()
plt.show()

print("âœ… Step 7B: Overlayed Histogram Graph Displayed")
```
